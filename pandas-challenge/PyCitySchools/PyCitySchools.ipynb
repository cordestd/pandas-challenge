```python
#Import all necessary libraries and tools
import pandas as pd
from pathlib import Path
```

```python
schools_complete = Path("../Starter_Code/PyCitySchools/Resources/schools_complete.csv")
students_complete = Path("../Starter_Code/PyCitySchools/Resources/students_complete.csv")
```

```python
school_data = pd.read_csv(schools_complete)
student_data = pd.read_csv(students_complete)

#I ran lines below to ensure the csv files were properly imported
#student_data.head()
#school_data.head()
```

```python
total_data = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
total_data.head()
```

```python
unique_schools = len(pd.unique(total_data["school_name"]))
unique_schools
```

```python
student_count = len(pd.unique(total_data["Student ID"]))
student_count
```

```python
budgets = pd.unique(total_data["budget"])
sum_budgets = budgets.sum()
sum_budgets
```

```python
math_mean = total_data["math_score"].mean()
math_mean
```

```python
reading_mean = total_data["reading_score"].mean()
reading_mean
```

```python
passing_math_count = total_data[(total_data["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage
```

```python
passing_reading_count = total_data[(total_data["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100
passing_reading_percentage
```

```python
passing_math_reading_count = total_data[
    (total_data["math_score"] >= 70) & (total_data["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
rough_summary = pd.Series([unique_schools, student_count, sum_budgets, math_mean, reading_mean, passing_math_percentage, passing_reading_percentage, overall_passing_rate], index = ["Total Schools", "Total Students", "Total Budget", "Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing"])

district_summary = pd.DataFrame(data=rough_summary)

district_summary = district_summary.transpose()

# Formatting
district_summary["Total Schools"] = district_summary["Total Schools"]
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

district_summary
```
